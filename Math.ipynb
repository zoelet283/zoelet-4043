# Plotting Functions

We can use the matplotlib.pyplot to plot functions

We will need to declare the x variable and a range of x values
Declare the y variable and assign values using the mathematical function

let's see this in practice


# y=sin(x)

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.arange(-2*np.pi, 2*np.pi, 0.2)

y=np.sin(x)

plt.plot(x,y)
plt.show()
```

# $ y = \sqrt(sin(x)) $

```python
x= np.arange(-2*np.pi, 2*np.pi, 0.01)

y=np.sqrt(np.sin(x))

plt.plot(x,y)
plt.show()
```

```python
x= np.arange(-2*np.pi, 2*np.pi, 0.01)

y=np.exp(-x)

plt.plot(x,y)
plt.show()
```

```python
x= np.arange(-2*np.pi, 2*np.pi, 0.01)

y=np.power(np.sin(x),2)

plt.plot(x,y)
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("$y=cos^2(x)$")
plt.show()
```

```python
x= np.arange(-2*np.pi, 2*np.pi, 0.01)

y=np.log(x)

plt.plot(x,y)
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("$y=ln(x)$")
plt.show()
```

# Exercise 2
Develop a Python function that accepts a quadratic equation and returns its roots. 

```python
def QuadRoots(a,b,c):
    x1=(-b+((b**2)-(4*a*c)))/(2*a)
    x2=(-b-((b**2)-(4*a*c)))/(2*a)
    x=[x1,x2]
    return x

x=QuadRoots(1,2,1)
print(x)
```

# Random Numbers

```python
# generate random floating point values
from random import seed
from random import random
# seed random number generator
seed(1)
# generate random numbers between 0-1
for i in range(1):
	value = random()
	print(value)
```

# Random Integer Number

```python
from random import seed
from random import randint
# seed random number generator
# seed(1)
# generate some integers
for _ in range(1):
	value = randint(100, 1000)
	print(value)
```

# Random Selection from a list 

```python
from random import seed
from random import choice
# seed random number generator
seed(1)
# prepare a sequence
sequence = [i for i in range(20)]
print(sequence)
# make choices from the sequence
for _ in range(5):
	selection = choice(sequence)
	print(selection)
```

# Random Sub-Sample

```python
from random import seed
from random import sample
# seed random number generator
seed(1)
# prepare a sequence
sequence = [i for i in range(20)]
print(sequence)
# select a subset without replacement
subset = sample(sequence, 5)
print(subset)
```

# Shuffle the list using NumPy

```python
# randomly shuffle a sequence
from numpy.random import seed
from numpy.random import shuffle
# seed random number generator
seed(1)
# prepare a sequence
sequence = [i for i in range(20)]
print(sequence)
# randomly shuffle the sequence
shuffle(sequence)
print(sequence)
```
